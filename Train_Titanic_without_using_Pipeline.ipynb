```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

**1.train_test_split from sklearn.model_selection:**

Purpose: Splits arrays or matrices into random train and test subsets.

Usage: Useful for creating training and testing datasets for model evaluation.

**2.SimpleImputer from sklearn.impute:**

Purpose: Handles missing data by replacing it with a specific value (mean, median, most frequent, or constant).

Usage: Essential for preprocessing datasets with missing values.

**3.OneHotEncoder from sklearn.preprocessing:**

Purpose: *Converts categorical variables into a format that can be provided to ML algorithms to do a better job in prediction*.

Usage: Transforms categorical data into a binary matrix.

**4.MinMaxScaler from sklearn.preprocessing:**

Purpose: Scales features to a given range, typically between zero and one.

Usage: Normalizes the range of independent variables or features of data.

**5.DecisionTreeClassifier from sklearn.tree:**

Purpose: A decision tree classifier.

Usage: Used for classification tasks, it splits the dataset into subsets based on the value of input features.

```python
# Calling the Train DataSet for predictions
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

The inplace=True parameter in the df.drop() method means that the operation will be performed directly on the DataFrame df without creating a new DataFrame. In other words, it modifies the original DataFrame by dropping the specified columns ('passengerId', 'Name', 'Ticket', and 'Cabin') and does not return a new DataFrame.

```python
df.head()
```

```python
# Step 1 : train/test/split
X_train,X_test,Y_train,Y_test = train_test_split(df.drop(columns=['Survived']),
                                         df['Survived'],test_size=0.2,random_state=42)
```

The **random_state parameter** in the train_test_split function is used to ensure reproducibility of your results. When you set random_state to a specific integer (like 42 in your example), it ensures that the random splitting of the data into training and testing sets is done in the same way every time you run the code. This means that you will get the same training and testing sets each time you run the code, which is useful for debugging and comparing results.

```python
# Applying imputation

si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_embarked
```

```python
X_train_age
```

```python
# one hot encoding Sex and Embarked

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_embarked
```

```python
X_train_sex
```

```python
df.head(2)
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_test_transformed
```

```python
X_train_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,Y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

The **accuracy_score** function from the **sklearn.metrics** module is used to evaluate the accuracy of a classification model. It calculates the ratio of correctly predicted instances to the total instances in the dataset. In other words, it measures how often the classifier makes correct predictions.
